1. Demo of a line plot on a polar axis.

```python
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np

r = np.arange(0, 2, 0.01)

theta = 2 * np.pi * r

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, r)
ax.set_rmax(3) #Maximum plot is set to 3
ax.set_rticks([1, 1.5, 2,2.7,3])  # Less radial ticks
ax.set_rlabel_position(-22.5)  # Move radial labels away from plotted line
ax.grid(True)

ax.set_title("A line plot on a polar axis", va='bottom')
plt.show()
```

The above polar plot shows polar coordinates. The maximum plot set is 3 and the theta is calculated on the basis of radial ticks. Basically, a polar plot can be used to avoid collisions between vessels and other ships or some naturally occured obstructions. It can be also used in radially symmetric wells to calculate the flow of ground water.
Reference: https://reference.com/history-geography/examples-real-life-uses-polar-coordinates-42f7df9b107292ae

2. The below given Seaborn example is a hexbin plot with marginal distributions

```python

import numpy as np
import seaborn as sns
sns.set_theme(style="ticks")

rs = np.random.RandomState(5)
x = rs.gamma(2, size=1000)
y = -.5 * x + rs.normal(size=1000)

sns.jointplot(x=x, y=y, kind="hex", color="#4CB391")
```

The RandomState value has been set to 5. Considering the data points, we can visualize the distribution of data. This is the best alternative to scatter plots when many points are overlapping. We can reveal patterns and trends
Ref: https://levelup.gitconnected.com/mastering-hexbin-plotting-in-python-a-beginners-guide-3626e0389c37

3. Population of Asian Continent

```python
import plotly.express as px
df = px.data.gapminder().query("year == 2007").query("continent == 'Asia'")
df.loc[df['pop'] < 2.e6, 'country'] = 'Other countries' # Represent only large countries
fig = px.pie(df, values='pop', names='country', title='Population of Asia continent')
fig.show()
```

The above chart shows Population of Asia continent according to the countries as per the year 2007. The data in px.pie is visualized by the sectors of pie which is set in values. Ref: https://plotly.com/python/pie-charts/

4. Plot of a pairwise data. (Tabular and Functional)

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# Making data
x = 0.5 + np.arange(8)
y = [1.8, 4.5, 2.5, 5, 6.5, 7.0, 2.6, 3.0]

# plot
fig, ax = plt.subplots()

ax.stem(x, y)

ax.set(xlim=(0, 5), xticks=np.arange(1, 10),
       ylim=(0, 5), yticks=np.arange(1, 10))

plt.show()
```

The above given Stem plot is like a histogram and is used to visualize data set and compare data.
Ref: https://statisticshowto.com/stemplot/
